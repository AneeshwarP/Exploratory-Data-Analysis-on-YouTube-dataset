# 5. Convert datatype of column published_at from object to pandas datetime.

```python
# Converting 'published_at' column to pandas datetime
data_Youtube['published_at'] = pd.to_datetime(data_Youtube['published_at'])

# Display the modified dataset
print(data_Youtube)
```

# 6.Create a new column as 'published_month' using the column published_at (display the months only)

```python
# Creating a new column 'published_month' with the months extracted from 'published_at'
data_Youtube['published_month'] = data_Youtube['published_at'].dt.month
print(data_Youtube)
```

# 7.Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb' and so on...

```python
# Creating & Mapping dictionary for month names for the column "published_month"
month_names = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
}

# Replace numbers with month names in the 'published_month' column
data_Youtube['published_month'] = data_Youtube['published_month'].replace(month_names)

# Display the modified dataset
print(data_Youtube)
```

```python
#Post Verfication : for Column "published_month"
data_Youtube["published_month"]
```

# 8.Find the number of videos published each month and arrange the months in a decreasing order based on the video count.

```python
# Using Group by 'published_month' and count() to find the number of videos in each month
video_counts_per_month = data_Youtube.groupby('published_month').size()

# Sorting the result in decreasing order based on video count
sorted_video_counts = video_counts_per_month.sort_values(ascending=True)

# printing the sorted result
print(sorted_video_counts)
```

# 10. Find the top10 channel names having the highest number of videos in the dataset and the bottom10 having lowest number of videos.

```python
data_Youtube.dtypes
```

```python
# Grouping by 'channel_title' and counting the number of videos for each channel
channel_video_counts = data_Youtube['channel_title'].value_counts()

# the top 10 channels with the highest number of videos
top10_channels = channel_video_counts.head(10)

# the bottom 10 channels with the lowest number of videos
bottom10_channels = channel_video_counts.tail(10)

# printing the top 10 and bottom 10 channel names
print("Top 10 Channels with Highest Number of Videos from Given Dataset:")
print(top10_channels.sort_values(ascending=False))

print("\nBottom 10 Channels with Lowest Number of Videos from Given Dataset:")
print(bottom10_channels)
```

# 13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.

```python
# Calculate the correlation between 'view_count' and 'dislikes'
correlation = data_Youtube['view_count'].corr(data_Youtube['dislikes'])

# Display the correlation value
print("Correlation between View Count and Dislikes:", round(correlation,2))

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='view_count', y='dislikes', data=data_Youtube)#plotting scatter plot 
sns.regplot(x='view_count', y='dislikes', data=data_Youtube)#plotting trend line  
plt.title("Scatter Plot: View Count vs Dislikes")
plt.xlabel("View Count")
plt.ylabel("Dislikes")
plt.show()
```

As per the above observation of the scatter plot & trend line for view count & the dislike we can say higher view counts as both higher or lower dislikes similarly for the Likes as well in vise versa 

hence as per the the correlation we can say it is greater than zero which is 0.68 & close to 1 by this we can say as view count increase there is drastic increase in dislikes for the video titles
